```python
import pickle
import itertools
import numpy as np
from sklearn import metrics
import matplotlib.pyplot as plt
import os
%matplotlib inline  
directorys=['real_plot_figures']  # temp models
for directory in directorys:
    if not os.path.exists(directory):
        os.makedirs(directory)
aucs=[]
p=10
# gound_truth_tsang

features_dic={}
features_dic['0']=['climbRate','Sgz','p','q ','curRoll','absRoll','diffClb','diffRollRate','diffDiffClb ','SaTime1','SaTime2','SaTime3','SaTime4','diffSaTime1','diffSaTime2','diffSaTime3','diffSaTime4', 'Sa']

features_dic['1']=['subject#','age','sex','test_time','Jitter(%)','Jitter(Abs)','Jitter:RAP','Jitter:PPQ5','Jitter:DDP','Shimmer','Shimmer(dB)','Shimmer:APQ3','Shimmer:APQ5','Shimmer:APQ11','Shimmer:DDA','NHR','HNR','RPDE','DFA','PPE']
#[str(x) for x in range(20)]
features_dic['2']=['id','Age','HoursPerWeek','TotalHours','APM','SelectByHotkeys','AssignToHotkeys','UniqueHotkeys','MinimapAttacks','MinimapRightClicks','NumberOfPACs','GapBetweenPACs','ActionLatency','ActionsInPAC','TotalMapExplored','WorkersMade','UniqueUnitsMade','ComplexUnitsMade','ComplexAbilitiesUsed']

#features_dic['2']=[str(x) for x in range(19)]#[]
features_dic['3']=['season','yr','mnth','hr','holiday','weekday','workingday','temp','atemp','hum','windspeed','weathersit_1','weathersit_2','weathersit_3','weathersit_4']
features_dic['4']=['longitude','latitude','housingMedianAge','totalRooms','totalBedrooms','population','households','medianIncome']
titlelist=['Elevators','Parkinsons','SkillCraft','Bike sharing','Cal housing']

for seed in range(5):
    for datasetindex in range(5):
        fr = open('UCB/'+str(datasetindex)+'.csv'+str(seed)+'UCB.txt','rb')
        summary=(pickle.load(fr))
        fr.close()
        plt.scatter(range(len(summary['ucb'])),-summary['ucb'])
        plt.scatter(range(len(summary['ucb'])),-summary['estimate'])
        plt.scatter(range(len(summary['ucb'])),-summary['lcb'])
        plt.title("Confidence interval for each arm")
        plt.xlabel("Arms' index")
        plt.savefig('real_plot_figures/UCBrealdata'+str(datasetindex+1)+'_seed_'+str(seed)+'.pdf',bbox_inches='tight',dpi='figure',pad_inches=0.1)

        p=int((np.sqrt(8*len(summary['ucb'])+1)+1)/2)
        Larms=[]
        for i in itertools.combinations(range(p),2):
            Larms.append(i)

        # print top k arms
        for i in np.argpartition(summary['ucb'], range(15))[:15]:
            print(Larms[i],summary['ucb'][i])

        Hessian=np.zeros([p,p])
        for i in range(len(Larms)):
            Hessian[Larms[i][0],Larms[i][1]]=abs(summary['estimate'][i])
        Hessian=Hessian.T
        
        fig, ax = plt.subplots()
        im = ax.imshow(Hessian,cmap=plt.cm.Blues)

        ax.set_xticks(np.arange(len(features_dic[str(datasetindex)])))
        ax.set_yticks(np.arange(len(features_dic[str(datasetindex)])))
        # ... and label them with the respective list entries
        ax.set_xticklabels(features_dic[str(datasetindex)])
        ax.set_yticklabels(features_dic[str(datasetindex)])
        ax.set_title(titlelist[datasetindex])
        # Rotate the tick labels and set their alignment.
        
        #plt.setp(ax.get_yticklabels(), va="center",
        # )
        plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
                 rotation_mode="anchor")

        fig.tight_layout()
        fig.savefig('real_plot_figures/realdata'+str(datasetindex+1)+'_seed_'+str(seed)+'.pdf',bbox_inches='tight',dpi=fig.dpi,pad_inches=0.1)
        plt.show()



```
